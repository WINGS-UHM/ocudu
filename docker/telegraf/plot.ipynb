```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
from scipy.signal import detrend

with open('csi_grid.pkl', 'rb') as f:
    data = pickle.load(f)

# get max number active REs
rows = 0
for d in data:
    if len(d) > rows:
        rows = len(d)

#  get number of columns (time samples)
cols = len(data) 
len(data)
```

```python


csi_grid = np.zeros((rows, cols), dtype=np.complex64)
```

```python
for i in range(cols):
    csi_grid[:len(data[i]), i] = data[i]

csi_grid = np.array(csi_grid)
csi_grid.shape
```

```python
# from scipy.io import savemat
# savemat('csi_grid.mat', {'csi_grid': csi_grid})
```

```python

plt.figure(figsize=(10, 6))
for i in range(rows):
    plt.plot(csi_grid[i, :].real)
plt.title('Magnitude of CSI Grid')
plt.xlabel('Time Index')
plt.ylabel('Magnitude')
plt.grid()
```

```python
# phase plot
phase_grid = detrend(np.unwrap(np.angle(csi_grid)), axis=1)
plt.figure(figsize=(10, 6))
for i in range(rows):
    plt.plot(phase_grid[i, :])
plt.title('Phase of CSI Grid')
plt.xlabel('Time Index')
plt.ylabel('Phase (radians)')
plt.grid()
plt.show()
```
